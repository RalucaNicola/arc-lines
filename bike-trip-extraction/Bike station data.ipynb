```python
import pandas as pd
import numpy as np
```

#### Load file

```python
df  = pd.read_csv(r".\bike_station_information.csv")
df
```

#### Data cleaning

- Filter only the Cambridge stations
- Make sure field types are correct
- Remove fields that are not needed

```python
# Filter Cambridge Stations
df = df[df["region_id"] == 8].reset_index(drop=True)
df
```

```python
# Drop columns Public and District
df.drop(columns=['capacity', 'has_kiosk', 'rental_methods', 'eightd_has_key_dispenser', 'region_id', 'external_id', 'electric_bike_surcharge_waiver', 'station_id', 'eightd_station_services', 'legacy_id', 'station_type'], axis=1, inplace=True)
df
```

```python
df.rename(columns={'short_name': 'Number', 'name': 'Name', 'lat': 'Latitude', 'lon': 'Longitude', 'capacity': 'Capacity', 'rental_methods': 'Rental methods', 'eightd_has_key_dispenser': 'Key Dispenser', 'has_kiosk': 'Kiosk'}, inplace=True)
df
```

```python
# Reorder columns
df = df[['Number', 'Name', 'Latitude', 'Longitude']].copy()
df
```

```python
np.random.seed(42)
random_bike_number = np.random.randint(0, 5, size=len(df))
random_bike_number
```

```python
df['InitialBikeCount'] = df['Capacity'] - random_bike_number
df
```

#### Export file

```python
df.to_csv(r".\bike_station_information_cambridge.csv", index=False)
```
